```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv',encoding = "utf-8")
df.head()
```

```python
for i in range(len(df['Topics'])):
    df['Topics'][i] = df['Topics'][i][:-1]
df.head()
```

```python
topics = set()
for i in df['Topics']:
    l = i.split('|')
    for topic in l:
        topics.add(topic)
topics
```

```python
problems = {}
difficulties = ['Easy', 'Medium', 'Hard']
for topic in topics:
    problems[topic] = {}
    for difficulty in difficulties:
        problems[topic][difficulty] = []
for i,problem in enumerate(df['Link']):
    tl = df['Topics'][i].split('|')
    for t in tl:
        if df['Difficulty'][i] in difficulties:
            problems[t][df['Difficulty'][i]].append(problem)
```

```python
link_topics = {}
for i,link in enumerate(df['Link']):
    link_topics[link] = {}
    link_topics[link]['Difficulty'] = df['Difficulty'][i]
    link_topics[link]['Topics'] = df['Topics'][i].split('|')
```

```python
users = pd.read_csv('users.csv',encoding='utf-8')
users.head()
```

```python
topic_points = {}
problems_done = {}
for topic in topics:
    topic_points[topic] = 0
    problems_done[topic] = {}
    for difficulty in difficulties:
        problems_done[topic][difficulty] = []
for problem in users['Easy'][0].split('|'):
    if(problem in link_topics):
        problem_topics = link_topics[problem]['Topics']
        for problem_topic in problem_topics:
            topic_points[problem_topic] += 1
            problems_done[problem_topic]['Easy'].append(problem)
for problem in users['Medium'][0].split('|'):
    if(problem in link_topics):
        problem_topics = link_topics[problem]['Topics']
        for problem_topic in problem_topics:
            topic_points[problem_topic] += 7
            problems_done[problem_topic]['Medium'].append(problem)
for problem in users['Hard'][0].split('|'):
    if(problem in link_topics):
        problem_topics = link_topics[problem]['Topics']
        for problem_topic in problem_topics:
            topic_points[problem_topic] += 20
            problems_done[problem_topic]['Hard'].append(problem)
problems_done
```

```python
total_points = {}
for topic in topics:
    total_points[topic] = 0
for topic in problems:
    mul = [1,7,20]
    for i,difficulty in enumerate(problems[topic]):
        total_points[topic] += len(problems[topic][difficulty])*mul[i]
#         total_points[topic] = 1
```

```python
suggestion={}
for i in total_points:
    if total_points[i] > 20:
        percentage = topic_points[i]*100/total_points[i]
        if(percentage < 50):
            suggestion[i] = []
            for difficulty in difficulties:
                count = 0
                for problem in problems[i][difficulty]:
                    if not(problem in problems_done[i][difficulty]) and count < 3:
                        suggestion[i].append(problem)
                        count += 1
suggestion
```
